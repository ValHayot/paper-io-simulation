This is the notebook to analyze the results of multi-threaded experiments. The log files should be generated with the instructions in README.md

Figure is generated and saved in `figures` directory.

```python
import matplotlib.pyplot as plt
import pandas as pd
import process_result
```

```python
# Ploting function with subplots
def plot_result():
    # Aggregate results in multiple files into one file.     
    process_result.export_real_results("local/real/", "aggregated.csv")
    process_result.export_simgrid_result("local/wrench/original/", "aggregated.csv")
    process_result.export_simgrid_result("local/wrench/pagecache/", "aggregated.csv")
    
    # Plot settings     
    plt.rcParams.update({'font.size': 8})
    fig, (ax1, ax2) = plt.subplots(figsize=(10, 5), ncols=2, nrows=1)

    process_result.plot_prop(ax1, "local", "readtime", "average read time")
    process_result.plot_prop(ax2, "local", "writetime", "average write time")

    plt.legend(loc='upper center', bbox_to_anchor=(-0.2, 1.3), ncol=3)
    plt.subplots_adjust(left=0.1, bottom=0.1, right=0.95, top=0.7, wspace=0.3)

    plt.savefig("figures/multi_local.pdf", format="pdf")
    plt.savefig("figures/multi_local.svg", format="svg")

    plt.show()
```

```python
plot_result()
```
